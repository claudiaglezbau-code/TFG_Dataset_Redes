```python
"""
MODELO DE HESTON — TFG — V3 (FINAL)
Fórmula Cerrada + Carr-Madan FFT + Red Neuronal  (4 x 256)
"""

import numpy as np
from scipy import fftpack, integrate
from math import pi
import csv, time
import pandas as pd
import matplotlib.pyplot as plt
import joblib

import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Constantes

S0  = 100.0
r   = 0.0
EPS = 1e-12

STRIKES_PCT = np.array([0.80, 0.85, 0.90, 0.95, 0.97, 0.99,
                         1.00, 1.01, 1.02, 1.05, 1.10, 1.15, 1.20])
MATURITIES  = np.array([0.1, 0.25, 0.5, 1.0])

TARGET_FILAS          = 400_000
PUNTOS_POR_SUPERFICIE = len(STRIKES_PCT) * len(MATURITIES)        # 52
N_SUPERFICIES         = int(np.ceil(TARGET_FILAS / PUNTOS_POR_SUPERFICIE))  # 7693
OUTFILE               = "dataset_heston_400k.csv"

SEED      = 42
TEST_SIZE = 0.15
VAL_SIZE  = 0.15
PARAMS_COLS = ['kappa', 'theta', 'sigma', 'rho', 'v0']


# BLOQUE 1 — FFunción característica de Heston

def caracteristica_heston(phi, params, S0_local, r_local, T, Pnum=2, unwrap_log=False):
    kappa, theta, sigma, rho, v0 = params
    phi    = np.array(phi, dtype=complex)
    i      = 1j
    u      = 0.5 if Pnum == 1 else -0.5
    b      = (kappa - rho * sigma) if Pnum == 1 else kappa
    a      = kappa * theta
    sigma2 = max(sigma**2, EPS)

    alpha = -0.5 * (phi**2) + i * u * phi
    beta  = b - rho * sigma * i * phi
    d     = np.sqrt(beta**2 - 2.0 * sigma2 * alpha)
    g     = (beta - d) / (beta + d)

    exp_dt         = np.exp(-d * T)
    one_minus_gexp = 1.0 - g * exp_dt
    one_minus_g    = 1.0 - g

    one_minus_gexp = np.where(np.abs(one_minus_gexp) < EPS, EPS, one_minus_gexp)
    one_minus_g    = np.where(np.abs(one_minus_g)    < EPS, EPS, one_minus_g)

    log_term = np.log(one_minus_gexp / one_minus_g)

    # Lord-Kahl (2010): el logaritmo complejo puede saltar de rama (±2πi)
    # conforme phi recorre la rejilla FFT. np.unwrap corrige esos saltos
    # forzando continuidad en la parte imaginaria.
    if unwrap_log and phi.ndim > 0 and len(phi) > 1:
        log_term = log_term.real + 1j * np.unwrap(log_term.imag)

    C = (1j * phi * r_local * T) + (a / sigma2) * ((beta - d) * T - 2.0 * log_term)
    D = ((beta - d) / sigma2) * (1.0 - exp_dt) / one_minus_gexp

    return np.exp(C + D * v0 + 1j * phi * np.log(S0_local))

def _integrando(phi, params, S0_local, r_local, T, K, Pnum):
    cf  = caracteristica_heston(phi, params, S0_local, r_local, T, Pnum)
    val = np.exp(-1j * phi * np.log(K)) * cf / (1j * phi)
    return np.real(val)


def heston_precio_cerrado(params, S0_local, r_local, T, K, tol=1e-6, limit=200):
    I2, _ = integrate.quad(_integrando, 0, np.inf,
                            args=(params, S0_local, r_local, T, K, 2),
                            limit=limit, epsabs=tol, epsrel=tol)
    P2 = 0.5 + I2 / pi

    I1, _ = integrate.quad(_integrando, 0, np.inf,
                            args=(params, S0_local, r_local, T, K, 1),
                            limit=limit, epsabs=tol, epsrel=tol)
    P1 = 0.5 + I1 / pi

    return max(0.0, S0_local * P1 - K * np.exp(-r_local * T) * P2)


def verificar_formula_cerrada(params, T=0.5, K=100.0):
    precio_cerrado = heston_precio_cerrado(params, S0, r, T, K)
    K_fft, P_fft   = carr_madan_fft(params, S0, r, T)
    precio_cm      = float(np.interp(K, K_fft, P_fft))

    print(f"\n  Heston 1993 (integración directa) : {precio_cerrado:.6f} €")
    print(f"  Carr-Madan FFT                    : {precio_cm:.6f} €")
    print(f"  Diferencia                        : {abs(precio_cerrado - precio_cm):.2e} €")
    return precio_cerrado, precio_cm

# BLOQUE 3 — Carr-Madam FFT

def carr_madan_fft(params, S0_local, r_local, T, alpha=1.5, N=4096, eta=0.2):
    i      = 1j
    lambd  = 2 * pi / (N * eta)
    b      = N * lambd / 2

    v      = np.arange(N) * eta
    phi    = v - (alpha + 1) * i

    char_vals  = caracteristica_heston(phi, params, S0_local, r_local, T, Pnum=2, unwrap_log=True)
    numerator  = np.exp(-r_local * T) * char_vals * np.exp(i * b * v)
    denom      = (alpha + i * v) * (alpha + i * v + 1)

    # ── Pesos de Simpson (Carr-Madan 1999) ──────────────────
    j = np.arange(N)
    w = 3 + (-1)**(j + 1)
    w[0] = 1

    fft_vals   = fftpack.fft((numerator / denom) * (eta / 3) * w).real

    k       = -b + j * lambd
    K_grid  = np.exp(k)
    precios = np.exp(-alpha * k) / pi * fft_vals

    return K_grid, precios

# BLOQUE 4 — Generación del dataset

def generar_paramsets_validos(n_needed, seed=SEED):
    rng        = np.random.default_rng(seed)
    valid_sets = []
    intentos   = 0
    while len(valid_sets) < n_needed:
        intentos += 1
        k   = rng.uniform(0.5, 5.0)
        t   = rng.uniform(0.01, 0.15)
        s   = rng.uniform(0.01, 0.6)
        rho = rng.uniform(-0.9, -0.1)
        v0  = rng.uniform(0.01, 0.2)
        if 2 * k * t > s**2:
            valid_sets.append([k, t, s, rho, v0])

    tasa = len(valid_sets) / intentos * 100
    print(f"  Parámetros válidos: {n_needed} de {intentos} intentos ({tasa:.1f}% aceptación)")
    return np.array(valid_sets)


def crear_dataset(n_superficies, outfile=OUTFILE):
    print(f"\n  Superficies : {n_superficies:,}  |  Puntos/sup: {PUNTOS_POR_SUPERFICIE}  |  Objetivo: {n_superficies * PUNTOS_POR_SUPERFICIE:,} filas")

    paramsets   = generar_paramsets_validos(n_superficies)
    strike_grid = S0 * STRIKES_PCT
    header      = ["kappa", "theta", "sigma", "rho", "v0", "T", "K", "price", "price_noisy"]
    rng         = np.random.default_rng(SEED)
    written     = 0
    errores     = 0
    t_start     = time.time()

    with open(outfile, mode='w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(header)

        for idx, params in enumerate(paramsets):
            if (idx + 1) % 500 == 0:
                elapsed = time.time() - t_start
                rate    = written / elapsed if elapsed > 0 else 1
                eta_s   = (TARGET_FILAS - written) / rate
                print(f"  [{idx+1:>5}/{n_superficies}]  {written:>8,} filas  "
                      f"({rate:.0f} filas/s)  ETA: {eta_s:.0f}s")

            for T in MATURITIES:
                try:
                    K_fft, P_fft   = carr_madan_fft(params, S0, r, T)
                    precios_interp = np.interp(strike_grid, K_fft, P_fft)
                    for j, K_val in enumerate(strike_grid):
                        base  = max(0.0, float(precios_interp[j]))
                        noisy = max(0.0, base * (1.0 + rng.normal(0.0, 0.01)))
                        writer.writerow([*params, float(T), float(K_val), base, noisy])
                        written += 1
                except Exception:
                    errores += 1
                    continue

    print(f"\n  Archivo   : {outfile}")
    print(f"  Filas     : {written:,}")
    print(f"  Errores   : {errores}")
    print(f"  Tiempo    : {(time.time()-t_start):.1f} s")
    return pd.read_csv(outfile)


# BLOQUE 5 — Validación deel dataset

def validar_dataset(df):
    feller      = (2 * df['kappa'] * df['theta'] > df['sigma']**2).all()
    intrinseco  = (S0 - df['K']).clip(lower=0)
    violaciones = (df['price'] < intrinseco - 1e-5).sum()
    precios_neg = (df['price'] < 0).sum()

    print(f"\n  Filas totales    : {len(df):,}")
    print(f"  Paramsets únicos : {df[PARAMS_COLS].drop_duplicates().shape[0]:,}")
    print(f"  Feller cumplido  : {feller}  {'✓' if feller else '✗ REVISAR'}")
    print(f"  Precios negativos: {precios_neg}  {'✓' if precios_neg == 0 else '✗ REVISAR'}")
    print(f"  Violac. arbitraje: {violaciones}  {'✓' if violaciones == 0 else '✗ REVISAR'}")

    p0  = df.iloc[0][PARAMS_COLS].values
    sub = df[(df['kappa'] == p0[0]) & (df['theta'] == p0[1])]

    fig = plt.figure(figsize=(10, 6))
    ax  = fig.add_subplot(111, projection='3d')
    ax.plot_trisurf(sub['K'], sub['T'], sub['price'], cmap='viridis', alpha=0.85)
    ax.set_title(f"Superficie de Precios Heston (muestra)\n"
                 f"κ={p0[0]:.2f}, θ={p0[1]:.4f}, σ={p0[2]:.2f}, ρ={p0[3]:.2f}, v₀={p0[4]:.4f}")
    ax.set_xlabel("Strike K (€)")
    ax.set_ylabel("Vencimiento T (años)")
    ax.set_zlabel("Precio call (€)")
    plt.tight_layout()
    plt.savefig("superficie_heston.png", dpi=130)
    plt.show()

# BLOQUE 6 — Preparación de datos

def preparar_datos(df):
    X = pd.DataFrame()
    X['log_moneyness'] = np.log(df['K'] / S0)
    X['sqrt_T']        = np.sqrt(df['T'].clip(lower=1e-12))
    X['kappa']         = df['kappa']
    X['log_theta']     = np.log(df['theta'].clip(lower=1e-8))
    X['sigma']         = df['sigma']
    X['rho']           = df['rho']
    X['log_v0']        = np.log(df['v0'].clip(lower=1e-8))

    y = (df['price'] / S0).values

    paramsets = df[PARAMS_COLS].drop_duplicates().reset_index(drop=True)
    rng       = np.random.default_rng(SEED)
    perm      = rng.permutation(len(paramsets))
    n_test    = int(len(paramsets) * TEST_SIZE)
    n_val     = int(len(paramsets) * VAL_SIZE)

    ps_test  = paramsets.iloc[perm[:n_test]]
    ps_val   = paramsets.iloc[perm[n_test:n_test + n_val]]
    ps_train = paramsets.iloc[perm[n_test + n_val:]]

    def mascara(ps):
        m = df[PARAMS_COLS].merge(ps, on=PARAMS_COLS, how='left', indicator=True)
        return (m['_merge'] == 'both').values

    m_train = mascara(ps_train)
    m_val   = mascara(ps_val)
    m_test  = mascara(ps_test)

    X_vals = X.values
    X_train_raw, y_train = X_vals[m_train], y[m_train]
    X_val_raw,   y_val   = X_vals[m_val],   y[m_val]
    X_test_raw,  y_test  = X_vals[m_test],  y[m_test]

    scaler  = StandardScaler()
    X_train = scaler.fit_transform(X_train_raw)
    X_val   = scaler.transform(X_val_raw)
    X_test  = scaler.transform(X_test_raw)

    joblib.dump(scaler, 'escalador_heston_nn.pkl')

    print(f"\n  Train : {X_train.shape[0]:>8,} filas  ({len(ps_train)} superficies)")
    print(f"  Val   : {X_val.shape[0]:>8,} filas  ({len(ps_val)} superficies)")
    print(f"  Test  : {X_test.shape[0]:>8,} filas  ({len(ps_test)} superficies)")

    return X_train, X_val, X_test, y_train, y_val, y_test, scaler


# BLOQUE 7 — Red neuronal(4 × 256)

def construir_red(input_dim, neuronas=256, n_capas=4, dropout=0.05):
    inp = layers.Input(shape=(input_dim,), name="entrada")
    x   = inp

    for i in range(n_capas):
        x = layers.Dense(neuronas, kernel_initializer='he_normal', name=f"densa_{i+1}")(x)
        x = layers.BatchNormalization(name=f"bn_{i+1}")(x)
        x = layers.Activation('elu', name=f"elu_{i+1}")(x)
        x = layers.Dropout(dropout, name=f"dropout_{i+1}")(x)

    salida = layers.Dense(1, activation='linear', name="precio_norm")(x)
    model  = models.Model(inputs=inp, outputs=salida, name="HestonSurrogate_4x256")
    model.compile(optimizer=optimizers.Adam(learning_rate=0.001), loss='mse', metrics=['mae'])
    return model


def get_callbacks():
    return [
        callbacks.EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True, verbose=1),
        callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, min_lr=1e-6, verbose=1),
        callbacks.ModelCheckpoint(filepath='mejor_modelo_heston.keras', monitor='val_loss',
                                  save_best_only=True, verbose=0)
    ]


# BLOQUE 8 — Evaluación

def evaluar_modelo(model, X_val, y_val, X_test, y_test, history=None):
    # Calcular sesgo en validación y aplicar corrección
    y_pred_val  = model(X_val, training=False).numpy().flatten()
    sesgo       = float(np.mean(y_pred_val - y_val))
    print(f"\n  Sesgo detectado en val : {sesgo * S0:.4f} €  → aplicando corrección")

    y_pred = model(X_test, training=False).numpy().flatten() - sesgo
    r2     = r2_score(y_test, y_pred)
    mae    = mean_absolute_error(y_test, y_pred)
    rmse   = np.sqrt(mean_squared_error(y_test, y_pred))

    print(f"  R²    : {r2:.6f}  {'✓ Excelente' if r2 > 0.99 else '▲ Aceptable' if r2 > 0.97 else '✗ Revisar'}")
    print(f"  MAE   : {mae:.6f}  →  {mae * S0:.4f} € error medio")
    print(f"  RMSE  : {rmse:.6f}  →  {rmse * S0:.4f} € RMSE")

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle("Evaluación — Red Neuronal Heston (4 × 256)", fontsize=14, fontweight='bold')

    if history:
        ax = axes[0]
        ax.semilogy(history.history['loss'],     label='Train', color='steelblue')
        ax.semilogy(history.history['val_loss'], label='Val',   color='tomato', ls='--')
        ax.set_title("Curva de Aprendizaje")
        ax.set_xlabel("Época")
        ax.set_ylabel("Pérdida MSE (log)")
        ax.legend(); ax.grid(True, alpha=0.3)

    ax     = axes[1]
    sample = np.random.choice(len(y_test), min(3000, len(y_test)), replace=False)
    ax.scatter(y_test[sample] * S0, y_pred[sample] * S0, alpha=0.3, s=5, color='steelblue')
    lim = max(y_test.max(), y_pred.max()) * S0 * 1.05
    ax.plot([0, lim], [0, lim], 'r--', lw=1.5, label='Predicción perfecta')
    ax.set_title(f"Precio Real vs Predicho  (R² = {r2:.5f})")
    ax.set_xlabel("Precio Real (€)"); ax.set_ylabel("Precio Predicho (€)")
    ax.legend(); ax.grid(True, alpha=0.3)

    ax      = axes[2]
    errores = (y_pred - y_test) * S0
    ax.hist(errores, bins=80, color='steelblue', edgecolor='white', linewidth=0.3)
    ax.axvline(0, color='red', ls='--', lw=1.5, label='Error = 0')
    ax.axvline(errores.mean(), color='orange', ls='-', lw=1.5,
               label=f'Media = {errores.mean():.4f} €')
    ax.set_title("Distribución del Error (Predicho − Real)")
    ax.set_xlabel("Error (€)"); ax.set_ylabel("Frecuencia")
    ax.legend(); ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig("evaluacion_red_heston.png", dpi=150, bbox_inches='tight')
    plt.show()

    return {"R2": r2, "MAE_norm": mae, "RMSE_norm": rmse,
            "MAE_eur": mae * S0, "RMSE_eur": rmse * S0, "sesgo": sesgo}


# BLOQUE 9 — Inferencia (pricer rápido)

def compilar_inferencia(model):
    """
    Compila el forward pass con tf.function y firma fija.
    Primera llamada: ~1s (TF traza y compila el grafo).
    Siguientes:      <0.5ms (solo cómputo matricial puro).
    """
    @tf.function(input_signature=[tf.TensorSpec(shape=[1, 7], dtype=tf.float32)])
    def _inferir(x):
        return model(x, training=False)
    return _inferir


def predecir_precio(inferir_fn, scaler, kappa, theta, sigma, rho, v0, T, K, sesgo=0.0):
    x_raw = np.array([[
        np.log(K / S0),
        np.sqrt(max(T, 1e-12)),
        kappa,
        np.log(max(theta, 1e-8)),
        sigma,
        rho,
        np.log(max(v0, 1e-8))
    ]], dtype=np.float32)
    x_scaled = scaler.transform(x_raw).astype(np.float32)
    pred_norm = float(inferir_fn(x_scaled).numpy()[0, 0])
    return (pred_norm - sesgo) * S0


# EJECUCIÓN PRINCIPAL

if __name__ == "__main__":

    tf.random.set_seed(SEED)
    np.random.seed(SEED)
    t_inicio = time.time()

    print("═" * 55)
    print("  PASO 0 — Verificación fórmula cerrada vs Carr-Madan")
    print("═" * 55)
    params_ejemplo = [2.0, 0.04, 0.3, -0.7, 0.04]
    verificar_formula_cerrada(params_ejemplo, T=0.5, K=100.0)

    print("\n" + "═" * 55)
    print("  PASO 1 — Generación del dataset (400.000 filas)")
    print("═" * 55)
    df_final = crear_dataset(N_SUPERFICIES, outfile=OUTFILE)

    print("\n" + "═" * 55)
    print("  PASO 2 — Validación del dataset")
    print("═" * 55)
    validar_dataset(df_final)

    print("\n" + "═" * 55)
    print("  PASO 3 — Preparación de datos")
    print("═" * 55)
    X_train, X_val, X_test, y_train, y_val, y_test, scaler = preparar_datos(df_final)

    print("\n" + "═" * 55)
    print("  PASO 4 — Entrenamiento (4 capas × 256 neuronas)")
    print("═" * 55)
    model = construir_red(input_dim=X_train.shape[1])
    model.summary()

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=300,
        batch_size=512,
        callbacks=get_callbacks(),
        verbose=1
    )

    print("\n" + "═" * 55)
    print("  PASO 5 — Evaluación sobre test")
    print("═" * 55)
    metricas = evaluar_modelo(model, X_val, y_val, X_test, y_test, history=history)
    sesgo    = metricas["sesgo"]
    joblib.dump(sesgo, 'sesgo_heston_nn.pkl')

    model.save('modelo_heston_final.keras')
    print("\n  Guardado: modelo_heston_final.keras")
    print("  Guardado: escalador_heston_nn.pkl")
    print("  Guardado: sesgo_heston_nn.pkl")

    print("\n" + "═" * 55)
    print("  PASO 6 — Comparativa de velocidad: red vs fórmula cerrada")
    print("═" * 55)
    kappa, theta, sigma, rho, v0, T, K = 2.0, 0.04, 0.3, -0.7, 0.04, 0.5, 100.0
    N_REPS = 100

    # Heston exacto
    tiempos_exacto = []
    for _ in range(N_REPS):
        t0 = time.perf_counter()
        precio_exacto = heston_precio_cerrado([kappa, theta, sigma, rho, v0], S0, r, T, K)
        tiempos_exacto.append((time.perf_counter() - t0) * 1000)
    t_exacto = np.mean(tiempos_exacto)

    # Red neuronal: compilar grafo + warmup, luego medir
    print("  Compilando grafo tf.function (primera vez tarda ~1s)...")
    inferir_fn = compilar_inferencia(model)
    predecir_precio(inferir_fn, scaler, kappa, theta, sigma, rho, v0, T, K, sesgo)  # warmup
    tiempos_nn = []
    for _ in range(N_REPS):
        t0 = time.perf_counter()
        precio_nn = predecir_precio(inferir_fn, scaler, kappa, theta, sigma, rho, v0, T, K, sesgo)
        tiempos_nn.append((time.perf_counter() - t0) * 1000)
    t_nn = np.mean(tiempos_nn)

    aceleracion = t_exacto / t_nn
    print(f"  Heston exacto : {precio_exacto:.6f} €  [{t_exacto:.3f} ms  (media {N_REPS} llamadas)]")
    print(f"  Red neuronal  : {precio_nn:.6f} €  [{t_nn:.3f} ms  (media {N_REPS} llamadas)]")
    print(f"  Error abs.    : {abs(precio_exacto - precio_nn):.6f} €")
    print(f"  Aceleración   : ×{aceleracion:.1f} más rápido")
    print(f"\n  Tiempo total del proceso: {(time.time()-t_inicio)/60:.1f} min")
```

```python
from scipy.optimize import brentq
from scipy.stats import norm
import pandas as pd
import numpy as np

# ── Funciones BS ──────────────────────────────────────────
def bs_call(S, K, T, r, sigma):
    if T <= 0 or sigma <= 0:
        return max(S - K * np.exp(-r * T), 0.0)
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    return S*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)

def precio_a_vi(precio, S, K, T, r=0.0):
    precio = float(precio)
    intrinseco = max(S - K*np.exp(-r*T), 0.0)
    if precio <= intrinseco + 1e-8:
        return np.nan
    try:
        return brentq(lambda sig: bs_call(S, K, T, r, sig) - precio,
                      1e-6, 5.0, xtol=1e-6)
    except Exception:
        return np.nan

# ── Recrear la partición exacta (mismo SEED) ──────────────
PARAMS_COLS = ['kappa', 'theta', 'sigma', 'rho', 'v0']
SEED        = 42
TEST_SIZE   = 0.15

df = pd.read_csv('dataset_heston_400k.csv')

paramsets = df[PARAMS_COLS].drop_duplicates().reset_index(drop=True)
rng       = np.random.default_rng(SEED)
perm      = rng.permutation(len(paramsets))
n_test    = int(len(paramsets) * TEST_SIZE)
ps_test   = paramsets.iloc[perm[:n_test]]

def mascara(ps):
    m = df[PARAMS_COLS].merge(ps, on=PARAMS_COLS, how='left', indicator=True)
    return (m['_merge'] == 'both').values

df_test = df[mascara(ps_test)].copy()

# ── Reconstruir features y predicciones ───────────────────
X_raw = pd.DataFrame({
    'log_moneyness': np.log(df_test['K'] / S0),
    'sqrt_T':        np.sqrt(df_test['T'].clip(lower=1e-12)),
    'kappa':         df_test['kappa'],
    'log_theta':     np.log(df_test['theta'].clip(lower=1e-8)),
    'sigma':         df_test['sigma'],
    'rho':           df_test['rho'],
    'log_v0':        np.log(df_test['v0'].clip(lower=1e-8)),
})

X_test = scaler.transform(X_raw.values)          # scaler ya está en memoria
y_test = (df_test['price'] / S0).values
y_pred = model(X_test, training=False).numpy().flatten() - sesgo  # sesgo en memoria

# ── Convertir a Volatilidad Implícita ─────────────────────
print("Calculando VI real...  (puede tardar ~1-2 min)")
vi_real = np.array([precio_a_vi(p*S0, S0, K, T, r)
                    for p, K, T in zip(y_test, df_test['K'], df_test['T'])])

print("Calculando VI predicha...")
vi_pred = np.array([precio_a_vi(p*S0, S0, K, T, r)
                    for p, K, T in zip(y_pred, df_test['K'], df_test['T'])])

mask = ~(np.isnan(vi_real) | np.isnan(vi_pred))
mae_vi = np.mean(np.abs(vi_real[mask] - vi_pred[mask]))

print(f"\n  Puntos válidos       : {mask.sum():,} / {len(mask):,}")
print(f"  MAE en Vol. Implícita: {mae_vi:.4f}  ({mae_vi*100:.2f} puntos porcentuales)")
```
